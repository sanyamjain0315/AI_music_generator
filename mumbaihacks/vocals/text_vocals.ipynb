```python
uberduck_auth = ("pub_vrcxacmixgtnekjjdj", "pk_486e4e83-9ce9-4b54-b6b9-b2b14a8830ba")
```

```python
voicemodel_uuid = "bdfde229-946d-4b17-8cec-2bacaec1ccfe"
text = "I'm sitting here in the dark, All alone with my fears. They're creeping up on me, And I don't know what to do. Fears, fears, They're all around me. Fears, fears, They're closing in. I can't escape, I can't run away. I'm trapped in this nightmare, With my fears."
```

```python
audio_uuid = requests.post(
    "https://api.uberduck.ai/speak",
    json=dict(speech=text, voicemodel_uuid=voicemodel_uuid),
    auth=uberduck_auth,
).json()["uuid"]
```

```python
audio_uuid = requests.post(
    "https://api.uberduck.ai/speak",
    json=dict(speech=text, voicemodel_uuid=voicemodel_uuid, bpm=144),
    auth=uberduck_auth,
).json()["uuid"]
```

```python
from time import sleep

for t in range(10):
    sleep(1) # check status every second for 10 seconds.
    output = requests.get(
        "https://api.uberduck.ai/speak-status",
        params=dict(uuid=audio_uuid),
        auth=uberduck_auth,
    ).json()
    if "path" in output:
        audio_url = output["path"]
        break
```

```python
Audio(audio_url)
```

```python
response = requests.get(audio_url)
file_path = "audio.mp3"

with open(file_path, "wb") as file:
    file.write(response.content)
```
